#  Ridge Regression (L2)

Ridge punishes large coefficients by adding a squared penalty.

It doesn't set coefficients to zero — just makes them smaller and smoother.

    Loss=MSE+λ∑w^2
example:

>You’re writing an exam and teacher says:

>"Don’t write long, complicated answers. Keep them short and neat."

>Ridge does exactly that — keeps weights small.

✔️ When to use Ridge

* Many features

* Multicollinearity

* You want smooth, stable weights

* You don’t want any feature to disappear

code from scratch for n dimn

```python
class MeraRidge:
    
    def __init__(self,alpha=0.1):
        
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None
        
    def fit(self,X_train,y_train):
        
        X_train = np.insert(X_train,0,1,axis=1)
        I = np.identity(X_train.shape[1])
        I[0][0] = 0 #I don't know why they do that but if i don't perform this step results are silently different from actual class if 
        #some one know that why they do than let me know in comment.
        result = np.linalg.inv(np.dot(X_train.T,X_train) + self.alpha * I).dot(X_train.T).dot(y_train)
        self.intercept_ = result[0]
        self.coef_ = result[1:]
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
        
```

```python
reg = MeraRidge()
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
print(r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

code from scratch for gradient descent 

```python
class MereRidgeGD:

    def __init__(self,epochs,learning_rate,alpha):
        self.lr= learning_rate
        self.epochs= epochs
        self.alpha=alpha
        self.coef_=None
        self.intercept_ =None

    def fit(self,X_train,y_train):

        self.coef_ = np.ones(X_train.shape[1])
        self.intercept_ = 0 
        thetha = np.insert(self.coef_,0,self.intercept_)

        X_train = np.insert(X_train,0,1,axis=1)

        for i in range(self.epochs):
            thetha_der = np.dot(X_train.T,X_train).dot(thetha)- np.dot(X_train.T,y_train)+self.alpha*thetha
            thetha=thetha - self.lr*thetha_der

        self.coef_ = thetha[1:]
        self.intercept_ = thetha[0]

    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+ self.intercept_
```

# Lasso Regression (L1)

Lasso adds a penalty on absolute values of weights.

    Loss=MSE+λ∑∣w∣

This penalty is sharp → it can push some coefficients to exactly zero.

Meaning:
Lasso removes unimportant features → acts like feature selection.

example:

>You’re cleaning your room.
Instead of rearranging everything (Ridge), you throw away useless items.

>Lasso throws away unnecessary features.

✔️ When to use Lasso

* Many features

* Want automatic feature selection

* Some features are useless

```python
from sklearn.linear_model import Lasso
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

plt.scatter(X,y)

from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train,y_train)
print(reg.coef_)
print(reg.intercept_)
```

```python
alphas = [0,1,5,10,30]
plt.figure(figsize=(12,6))
plt.scatter(X,y)
for i in alphas:
    L = Lasso(alpha=i)
    L.fit(X_train,y_train)
    plt.plot(X_test,L.predict(X_test),label='alpha={}'.format(i))
plt.legend()
plt.show()
```

#  ElasticNet (L1 + L2):

*ElasticNet = Ridge + Lasso together.*

L1 removes irrelevant features

L2 stabilizes remaining features

Best of both worlds.

Equation:

     Loss=MSE+λ1​∑∣w∣+λ2​∑w^2

> Cleaning your house:
> 
> L1 → throw away useless stuff
> 
> L2 → organize remaining stuff neatlyElasticNet does both.


✔️ When to use ElasticNet

* When dataset is large

* When features are correlated

* When you want both selection + stability
